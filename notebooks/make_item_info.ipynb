### Item data 

```python
profit_item = pd.read_csv('../data/raw/Profit per Product.csv')
```

```python
profit_item['Item'].value_counts().shape
```

```python
profit_item = profit_item.iloc[:18,:2]
```

```python
profit_item
```

```python
Ind_item = pd.DataFrame(Industry[['Product Type', 'Item']].value_counts())
Ind_item.columns = ['Industry_count']
```

```python
Ind_item
```

```python
ret_item = pd.DataFrame(retail[['Product Type','Item']].value_counts())
```

```python
ret_item
```

```python
item_info = pd.concat([Ind_item, ret_item], axis =1)
item_info.columns = ['Industry_count', 'retail_count']
```

```python
item_info
```

```python
profit_item.index = profit_item['Item']
```

```python
profit_item.drop('Item', axis = 1, inplace=True)
```

```python
profit_item
```

```python
join_col = item_info.index.get_level_values('Item')
item_info['Profit'] = profit_item.loc[join_col].values
```

```python
item_info
```

```python
item_info.to_csv('../data/preprocessed/item_info.csv')
```
