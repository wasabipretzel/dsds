```python
dealer = pd.read_excel(r'C:/Users/dev_account/dev/after_univ/2022/hanyang_dsproject/data/raw/Dealer_Hierarchies.xlsx')
```

```python
dealer.head()
```

```python
pd.DataFrame(dealer['Region Level 1'].value_counts()).index.tolist()
```

```python
geo_coord = {}
geolocator = Nominatim(user_agent='jinwooahn')
address_list = pd.DataFrame(dealer['Region Level 1'].value_counts()).index.tolist()
for address in address_list:
    location = geolocator.geocode(address)
    if location == None:
        geo_coord[address] = [0, 0]
    else:
        geo_coord[address] = [location.latitude, location.longitude]
    time.sleep( random.uniform(1,3) )
```

```python
geo_coord
```

```python
pd.DataFrame(geo_coord).T.to_excel('C:/Users/dev_account/dev/after_univ/2022/hanyang_dsproject/data/preprocessed/region_data.xlsx')
```
