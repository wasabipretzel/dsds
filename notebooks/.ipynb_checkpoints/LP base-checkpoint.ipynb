## Preprocessing

```python
#1. Item Tag 만들기
item_list = profit.Item.tolist()
profit_list = profit.Profit.tolist()
item_tag = {}
for i, items in enumerate(item_list):
    item_tag[i] = [items,profit_list[i]]
print(item_tag)
```

```python
sub_item_list = ['Ear Buds','Mouse','Keyboard','Cereal','Shrimp','Noodles','Steak','Nails','Glue','Phone Charger',
           'Eggs','King Crab','Tape','Bracket','Paint','Power Cord','Milk']
```

```python
# Item tag with Item name
item_list = profit.Item.tolist()
item_tag2 = {}
for i, items in enumerate(item_list):
    item_tag2[items] = i
print(item_tag2)
```

```python
# industry_train.drop(['var'], axis = 1, inplace=True)
```

```python
industry_train
```

```python
def match(item):
    return item_tag2[item]
industry_train['Item_var'] = industry_train['Item'].apply(match)
```

```python
industry_train
```

```python
#2. 각 store의 item 변수 명을 "store1_{item_tag_num}" 으로 정의한다
industry_train['var'] = industry_train.apply(lambda row : str(row.Store) + "_" + str(row.Item_var), axis =1)
```

```python
industry_const.iloc[0]
```

```python
# industry_const을 dict 형태로 만들자. var -> industry size로
ind_const = {}
for i in range(industry_const.shape[0]):
    ind_const[industry_const.iloc[i][1]] = industry_const.iloc[i][0]
```

```python
ind_const
```

```python
temp_li = list(ind_const.keys())
for i in range(3):
    print(item_tag[int(temp_li[i].split('_')[1])][1])
```

```python
item_tag[int(list(ind_const.keys())[0].split('_')[1])]
```

```python
item_tag
```

```python
model = LpProblem(sense = LpMaximize)

# 변수는 각 store의 각 상품들.. 총 642 * 18
# sigma 각 변수 * profit -> obj
# constraint 각 변수 <= ind size -> 각 변수와 ind size을 두개 붙여놓은 df가 필요
# -> 2021년의 해당 df가 industry_const
x_name = industry_const['var'].tolist()
x = [LpVariable(x_name[i], lowBound=0, upBound = ind_const[x_name[i]]) for i in range(len(x_name))]
# obj function
# 1_10 * profit of 1_10 + ...
obj_func = [(x[i] ,item_tag[int(str(x[i]).split('_')[1])][1]) for i in range(len(x_name))]
obj_function = LpAffineExpression(obj_func)
```

```python
import numpy as np
```

```python
np.sum(x)
```

```python
# 전체 주어진 물량 개수는 N개
# 2021년 1월 retail의 물량은 1153
model += (np.sum(x) <= N)
model += obj_function
```

```python
model
```

```python
status = model.solve()
```

```python
print(f"status: {model.status}, {LpStatus[model.status]}")
print(f'objective : {model.objective.value()}')
```

```python
temp = []
for var in model.variables():
    temp.append([var.name, var.value()])
#     print(f"{var.name} : {var.value()}")
```

```python
df = pd.DataFrame(temp, columns = ['name','value'])
print(df)
```
