```python
# 예시를 위해 Store1의 특정 품목을 예측해보자
# 먼저, ARIMA를 사용할 수 있는지 Autocorrelation이 존재하는지 확인해보자
# data import
import pandas as pd
industry = pd.read_csv('../data/raw/Industry Data.csv')
retail = pd.read_csv('../data/raw/Retail Data.csv')

#store1 의 key board 데이터를 추출
store1 = industry[industry['Store'] == 1]
store1 = store1[store1['Item'] == 'Keyboard']
```

```python
# 균일성을 위해 매 월로 설정하되, 없는 달에는 0을 삽입한다
store1.Month = pd.to_datetime(store1.Month)
```

```python
store1 = store1[['Month', 'Industry Size']]
```

```python
# 2015년부터 2022-02까지.. 
df = pd.DataFrame({'Month' : pd.date_range(start='2015-01-01', end='2022-02-01', freq='MS'), 'Industry Size' : 0})
df
```

d = 2, p = 2, q = 2

```python
store1['Industry Size']
```

```python
train = store1['Industry Size'][:76]
test = store1['Industry Size'][76:]
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
arima_model =ARIMA(train, order = (2, 2, 2)) #pdq
model = arima_model.fit()
print(model.summary())
```

```python
y_pred = pd.Series(model.forecast(10))
y_true = test
```

```python
y_pred
```

```python
test
```

```python
y_pred.plot(color = 'blue')
y_true.plot(color = 'red')
```

```python
#쓰레기
```

```python
data = store1['Industry Size'].diff().dropna()
```

```python
data
```

```python
train = data[:76]
test = data[76:]
```

```python
train
```

```python
import pmdarima as pm
```

```python
model = pm.auto_arima(y = train, d = 1, start_p=0, max_p=3, start_q=0, max_q = 3, m = 1,
                     seasonal=True, stepwise=True, trace=True)
```

```python
model.fit(train)
```

```python
y_pred = pd.Series(model.predict(n_priods =len(test)))
```

```python
y_pred.plot(color = 'blue')
test.plot(color = 'red')
```
