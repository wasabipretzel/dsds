```python
import pandas as pd
industry = pd.read_csv('../data/raw/Industry Data.csv')
```

```python
industry
```

```python
industry.groupby('Item')['Industry Size'].sum()
```

```python
# store1 을 골라서, 매 월이 index가 되고, Item이 column이 되고, ind size가 value가 되게끔 df 만들자
store1 = industry[industry['Store'] == 1]
store1
```

```python
store1['Item'].value_counts().index.tolist()
```

```python
cols = store1['Item'].value_counts().index.tolist()
df = pd.DataFrame(0, columns = cols, index = pd.date_range('2015-01-01', '2022-03-01', freq='MS') )
```

```python
df
```

```python
df.loc['2015-01-01']['Shrimp']
```

```python
store1.index = pd.to_datetime(store1.Month)
```

```python
store1.Month = pd.to_datetime(store1.Month)
```

```python
store1
```

```python
store1.index
```

```python
store1.iloc[0]
```

```python
for i in range(store1.shape[0]):
    df.loc[store1.iloc[i]['Month']][store1.iloc[i]['Item']] = store1.iloc[i]['Industry Size']
```

```python
df
```

```python
df.corr()
```
